```
import matplotlib.pyplot as plt
import seaborn as sns
col_names = ContinuousVar
#col_names.extend(TargetVar)
fig, ax = plt.subplots(len(col_names),2, figsize=(14,60))

for i, col_val in enumerate(col_names):

    sns.distplot(edit[col_val].fillna(0), hist=True, ax=ax[i,0])
    ax[i,0].set_title('Freq dist '+col_val, fontsize=10)
    ax[i,0].set_xlabel(col_val, fontsize=8)
    ax[i,0].set_ylabel('Count', fontsize=8)
    sns.boxplot(x=edit[col_val], data=edit, palette="Set3", ax=ax[i,1])

plt.show()

plt.tight_layout(pad = 2.5)
```

```
import matplotlib.pyplot as plt
import seaborn as sns
col_names = ContinuousVar
#col_names.extend(TargetVar)
fig, ax = plt.subplots(len(col_names),3, figsize=(14,100))
plt.subplots_adjust(bottom=0.1)
for i, col_val in enumerate(col_names):
    #Distribution
    sns.distplot(edit.loc[edit['TARGET'] == 0, col_val].fillna(0), label = 'target == 0', ax=ax[i,0])
    # plot loans that were not repaid
    sns.distplot(edit.loc[edit['TARGET'] == 1, col_val].fillna(0), label = 'target == 1', ax=ax[i,0])    
    # Label the plots
    plt.title('Distribution of %s by Target Value' % col_val)
    plt.xlabel('%s' % col_val); plt.ylabel('Density');

    #Boxplot
    sns.boxplot(y=edit[col_val], x = edit['TARGET'], palette="Set3", ax=ax[i,1])

    #Nulls
    if(len(edit[edit[col_val].isnull()]))!=0:
      sns.countplot(data=edit[edit[col_val].isnull()], x = 'TARGET', ax=ax[i,2])

plt.show()

plt.tight_layout(pad = 2.5)
```

Check for distribution of NAs

```
print("Overall\n",edit.TARGET.value_counts()*100/len(edit))

# iterate through the new features
for i, feature in enumerate(ContinuousVar):
    x = edit[edit[feature].isnull()]
    if len(x)==0:
      print(feature+' has no nulls\n')
    elif len(x[x['TARGET']==0])==0:
      print(feature+' has no 0s in nulls\n')
    elif len(x[x['TARGET']==1])==0:
      print(feature+' has no 1s in nulls\n')
    else:
      print(feature, '\n',x.TARGET.value_counts()[0],'\n',(x.TARGET.value_counts()*100/len(x))[0] ,'\n')
      labels = '0','1'
      sizes = [(x.TARGET.value_counts()*100/len(x))[0],(x.TARGET.value_counts()*100/len(x))[1]]
      colors = ['lightcoral', 'lightskyblue']
      explode = (0.1, 0)  # explode 1st slice

      # Plot
      plt.pie(sizes, explode=explode, labels=labels, colors=colors,
      autopct='%1.1f%%', shadow=True, startangle=140)

      plt.axis('equal')
      plt.show()
```
